```python
import datetime as dt

# Convert date column to datetime (replace 'purchase_date' with the relevant column name)
# Handle non-date values by coercing errors
df['Stay_In_Current_City_Years'] = pd.to_datetime(df['Stay_In_Current_City_Years'], errors='coerce')

# Create 'recency' feature
now = dt.datetime.now()
# Fill missing values in 'Stay_In_Current_City_Years' with a placeholder (e.g., today's date) before calculating recency
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].fillna(now)
df['recency'] = (now - df['Stay_In_Current_City_Years']).dt.days

# Create 'frequency' feature
frequency = df.groupby('User_ID').size().reset_index(name='frequency')

# Create 'monetary' feature
monetary = df.groupby('User_ID')['Purchase'].sum().reset_index(name='monetary')

# Combine features
features = df[['User_ID', 'recency']].drop_duplicates().merge(frequency, on='User_ID').merge(monetary, on='User_ID')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features[['recency', 'frequency', 'monetary']])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
features['cluster'] = kmeans.fit_predict(features_scaled)
```

```python
# Display cluster characteristics
print(features.groupby('cluster').mean())
```

```python
sns.pairplot(features, hue='cluster', vars=['recency', 'frequency', 'monetary'])
plt.title('Customer Segments')
plt.show()
```
